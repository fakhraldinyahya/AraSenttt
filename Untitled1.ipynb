```python
# الخلية 2: إنشاء بيانات تدريب موسعة
import pandas as pd
import numpy as np
from datetime import datetime

# قوالب للجمل الإيجابية
positive_templates = [
    "الاكل لذيذ جدا", "الاكل رائع", "الاكل ممتاز", "الاكل طيب", "الاكل شهي",
    "الخدمة ممتازة", "الخدمة رائعة", "الخدمة سريعة", "الموظفين محترمين", "الاستقبال حلو",
    "السعر مناسب", "السعر معقول", "السعر ممتاز", "الاسعار مناسبة", "الاسعار رخيصة",
    "الجلسات روعة", "الجلسات مريحة", "الاضاءة حلوة", "المكان رايق", "الاجواء جميلة",
    "المطعم نظيف", "النظافة ممتازة", "المكان مرتب", "الاكل نظيف", "الادوات نظيفة",
    "الموقع سهل", "المواقف متوفرة", "الوصول سهل", "المكان قريب", "الموقع استراتيجي",
    "التوصيل سريع", "التوصيل ممتاز", "الطلب وصل بسرعة", "التغليف حلو", "الطلب حار"
]

# قوالب للجمل السلبية
negative_templates = [
    "الاكل سيء", "الاكل مو طيب", "الاكل بارد", "الاكل مالح", "الاكل مقزز",
    "الخدمة سيئة", "الخدمة بطيئة", "الموظفين غير محترمين", "الاستقبال سيء", "الخدمة زفت",
    "السعر غالي", "السعر مرتفع", "الاسعار غالية", "الاسعار مبالغ فيها", "السعر مو معقول",
    "الجلسات ضيقة", "الاضاءة سيئة", "المكان زحمة", "الاجواء مو حلوة", "المكان مزعج",
    "المطعم وسخ", "النظافة سيئة", "المكان غير نظيف", "الاكل غير نظيف", "الادوات وسخة",
    "الموقع صعب", "المواقف غير متوفرة", "الوصول صعب", "المكان بعيد", "الموقع سيء",
    "التوصيل متأخر", "التوصيل سيء", "الطلب جاء بارد", "الطلب ناقص", "الطلب غلط"
]

# قوالب للجمل المحايدة
neutral_templates = [
    "الاكل عادي", "الاكل متوسط", "الاكل مقبول", "الاكل مو مره", "الاكل عادي جدا",
    "الخدمة عادية", "الخدمة متوسطة", "الموظفين عاديين", "الاستقبال عادي", "الخدمة مقبولة",
    "السعر متوسط", "السعر مقبول", "الاسعار متوسطة", "السعر عادي", "الاسعار عادية",
    "الجلسات عادية", "الاضاءة عادية", "المكان عادي", "الاجواء عادية", "المكان هادي",
    "النظافة عادية", "المكان نظيف نوعا ما", "النظافة مقبولة", "الاكل نظيف", "النظافة مو مره",
    "الموقع متوسط", "المواقف متوسطة", "الوصول عادي", "المكان قريب نوعا ما", "الموقع مو بعيد",
    "التوصيل متوسط", "التوصيل عادي", "الطلب وصل عادي", "التغليف عادي", "الطلب وصل في الوقت"
]

# توليد البيانات
data = []
for i in range(100):  # 100 عينة إيجابية
    for template in np.random.choice(positive_templates, 1):
        data.append({"text": template, "sentiment": "positive"})

for i in range(70):  # 70 عينة سلبية
    for template in np.random.choice(negative_templates, 1):
        data.append({"text": template, "sentiment": "negative"})

for i in range(30):  # 30 عينة محايدة
    for template in np.random.choice(neutral_templates, 1):
        data.append({"text": template, "sentiment": "neutral"})

# خلط البيانات
np.random.shuffle(data)

# إنشاء DataFrame
df = pd.DataFrame(data)

# حفظ البيانات إلى ملف Excel
df.to_excel('training_data.xlsx', index=False)
print(f"✅ تم إنشاء {len(df)} عينة تدريب")
print("\n📊 توزيع المشاعر:")
print(df['sentiment'].value_counts())
print("\n📊 أول 10 عينات:")
print(df.head(10))
```

```python
# الخلية 3: تحميل البيانات وتجهيز النموذج
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
from sklearn.model_selection import train_test_split
import os

# تحديد المسار مع اسم النموذج والتاريخ
timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
MODEL_PATH = f"E:/project2026/projectfood/AraSenttt/models/sentiment_model_{timestamp}"
os.makedirs(MODEL_PATH, exist_ok=True)
print(f"📁 سيتم حفظ النموذج في: {MODEL_PATH}")

# تحميل البيانات
print("📂 جاري تحميل البيانات...")
df = pd.read_excel('training_data.xlsx')
print(f"✅ تم تحميل {len(df)} عينة")

# تحويل المشاعر إلى أرقام
sentiment_map = {'positive': 0, 'negative': 1, 'neutral': 2}
df['label'] = df['sentiment'].map(sentiment_map)

# تقسيم البيانات
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42, stratify=df['label'])
print(f"\n📊 بيانات التدريب: {len(train_df)}")
print(f"📊 بيانات التحقق: {len(val_df)}")

# تحميل tokenizer
print("\n📥 جاري تحميل tokenizer...")
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv2")

# تجهيز البيانات
def tokenize_function(examples):
    return tokenizer(examples['text'], padding='max_length', truncation=True, max_length=128)

train_dataset = Dataset.from_pandas(train_df[['text', 'label']])
val_dataset = Dataset.from_pandas(val_df[['text', 'label']])

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)

# تحميل النموذج
print("📥 جاري تحميل النموذج...")
model = AutoModelForSequenceClassification.from_pretrained(
    "aubmindlab/bert-base-arabertv2",
    num_labels=3,
    ignore_mismatched_sizes=True  # لتجاهل مشاكل الأوزان
)

print("✅ تم تجهيز النموذج والبيانات")
```

```python
# الخلية 4: التدريب والحفظ مع اسم النموذج
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

# دالة التقييم
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    accuracy = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average='weighted')
    precision = precision_score(labels, predictions, average='weighted', zero_division=0)
    recall = recall_score(labels, predictions, average='weighted', zero_division=0)
    return {
        'accuracy': accuracy,
        'f1': f1,
        'precision': precision,
        'recall': recall
    }

# إعداد التدريب
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=10,  # زيادة عدد الدورات
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    report_to="none",
    logging_steps=10,
    save_total_limit=2,  # حفظ آخر نموذجين فقط
    metric_for_best_model="accuracy",
    greater_is_better=True
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

# بدء التدريب
print("\n" + "="*60)
print("🚀 بدء تدريب النموذج...")
print("="*60)

trainer.train()

# حفظ النموذج مع التاريخ
print("\n" + "="*60)
print(f"💾 جاري حفظ النموذج في: {MODEL_PATH}")
print("="*60)

model.save_pretrained(MODEL_PATH)
tokenizer.save_pretrained(MODEL_PATH)

# حفظ معلومات النموذج في ملف نصي
with open(f"{MODEL_PATH}/model_info.txt", 'w', encoding='utf-8') as f:
    f.write(f"Model Name: sentiment_model_{timestamp}\n")
    f.write(f"Training Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
    f.write(f"Training Samples: {len(train_df)}\n")
    f.write(f"Validation Samples: {len(val_df)}\n")
    f.write(f"Epochs: 10\n")
    f.write(f"Batch Size: 8\n")

print("✅ تم حفظ النموذج بنجاح")

# تقييم نهائي
print("\n📊 نتائج التقييم النهائية:")
print("="*40)
results = trainer.evaluate()
for key, value in results.items():
    if key.startswith('eval_'):
        print(f"{key}: {value:.4f}")

# اختبار النموذج
print("\n🧪 اختبار النموذج:")
print("="*60)

test_texts = [
    "الاكل لذيذ جداً",
    "الخدمة سيئة",
    "السعر مناسب",
    "المطعم نظيف ومرتب",
    "التوصيل متأخر",
    "الاسعار غالية شوي",
    "الموظفين محترمين",
    "الجلسات ضيقة"
]

model.eval()
for text in test_texts:
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.softmax(outputs.logits, dim=1)
        predicted_class = torch.argmax(predictions, dim=1).item()
    
    sentiment = {0: 'إيجابي 😊', 1: 'سلبي 😞', 2: 'محايد 😐'}[predicted_class]
    confidence = predictions[0][predicted_class].item()
    
    print(f"\n📝 النص: {text}")
    print(f"🎯 المشاعر: {sentiment}")
    print(f"📊 الثقة: {confidence:.2%}")

# عرض مسار النموذج المحفوظ
print("\n" + "="*60)
print(f"✅ النموذج محفوظ في: {MODEL_PATH}")
print("="*60)

# نسخ النموذج إلى المسار الرئيسي (اختياري)
import shutil
MAIN_MODEL_PATH = "E:/project2026/projectfood/AraSenttt/models/sentiment_model"
if os.path.exists(MAIN_MODEL_PATH):
    shutil.rmtree(MAIN_MODEL_PATH)
shutil.copytree(MODEL_PATH, MAIN_MODEL_PATH)
print(f"✅ تم نسخ النموذج إلى: {MAIN_MODEL_PATH}")
```
