```python
# الخلية 2: تحميل البيانات من ملف Excel
import pandas as pd
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import Dataset
from sklearn.model_selection import train_test_split
import os

# تحديد المسارات
MODEL_PATH = "E:/project2026/projectfood/AraSenttt/models/sentiment_model"
os.makedirs(MODEL_PATH, exist_ok=True)

# تحميل البيانات من ملف Excel
print("📂 جاري تحميل البيانات من ملف Excel...")
df = pd.read_excel('training_data.xlsx')
print(f"✅ تم تحميل {len(df)} عينة")
print("\nتوزيع المشاعر:")
print(df['sentiment'].value_counts())

# تحويل المشاعر إلى أرقام
sentiment_map = {'positive': 0, 'negative': 1, 'neutral': 2}
df['label'] = df['sentiment'].map(sentiment_map)

# تقسيم البيانات
train_df, val_df = train_test_split(df, test_size=0.2, random_state=42)
print(f"\n📊 بيانات التدريب: {len(train_df)}")
print(f"📊 بيانات التحقق: {len(val_df)}")
```

```python
# الخلية 3: تحميل النموذج والتجهيز
print("📥 جاري تحميل النموذج...")

# تحميل tokenizer
tokenizer = AutoTokenizer.from_pretrained("aubmindlab/bert-base-arabertv2")

# تجهيز البيانات
def tokenize_function(examples):
    return tokenizer(examples['text'], padding='max_length', truncation=True, max_length=128)

train_dataset = Dataset.from_pandas(train_df[['text', 'label']])
val_dataset = Dataset.from_pandas(val_df[['text', 'label']])

train_dataset = train_dataset.map(tokenize_function, batched=True)
val_dataset = val_dataset.map(tokenize_function, batched=True)

# تحميل النموذج
model = AutoModelForSequenceClassification.from_pretrained(
    "aubmindlab/bert-base-arabertv2",
    num_labels=3
)

print("✅ تم تحميل النموذج والتجهيز")
```

```python
# الخلية 4: التدريب والحفظ
from sklearn.metrics import accuracy_score, f1_score

# دالة التقييم
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    accuracy = accuracy_score(labels, predictions)
    f1 = f1_score(labels, predictions, average='weighted')
    return {'accuracy': accuracy, 'f1': f1}

# إعداد التدريب
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    eval_strategy="epoch",
    save_strategy="epoch",
    load_best_model_at_end=True,
    report_to="none",
    logging_steps=10
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    compute_metrics=compute_metrics,
)

# بدء التدريب
print("🚀 بدء التدريب...")
print("="*50)
trainer.train()

# حفظ النموذج
print("\n💾 جاري حفظ النموذج...")
model.save_pretrained(MODEL_PATH)
tokenizer.save_pretrained(MODEL_PATH)
print(f"✅ تم حفظ النموذج في: {MODEL_PATH}")

# تقييم نهائي
print("\n📊 نتائج التقييم النهائية:")
results = trainer.evaluate()
for key, value in results.items():
    print(f"{key}: {value:.4f}")

# اختبار النموذج
print("\n🧪 اختبار النموذج:")
print("="*50)

test_texts = [
    "الاكل لذيذ جداً",
    "الخدمة سيئة",
    "السعر مناسب",
    "المطعم نظيف ومرتب",
    "التوصيل متأخر"
]

for text in test_texts:
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    
    with torch.no_grad():
        outputs = model(**inputs)
        predictions = torch.softmax(outputs.logits, dim=1)
        predicted_class = torch.argmax(predictions, dim=1).item()
    
    sentiment = {0: 'إيجابي 😊', 1: 'سلبي 😞', 2: 'محايد 😐'}[predicted_class]
    confidence = predictions[0][predicted_class].item()
    
    print(f"\n📝 النص: {text}")
    print(f"🎯 المشاعر: {sentiment}")
    print(f"📊 الثقة: {confidence:.2%}")
```
